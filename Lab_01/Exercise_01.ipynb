```python
import pandas as pd

# Load the California Housing dataset
df = pd.read_csv('/content/sample_data/california_housing_train.csv')

# Display the first few rows of the DataFrame to verify
print("California Housing Dataset loaded successfully. First 5 rows:")
df.head()
```

## Explore dataset.describe()

```python
print(df.describe())
```

## Display rows where 'total_bedrooms' is greater than 310

### Subtask:
Filter the DataFrame to display only the rows where the value in the 'total_bedrooms' column is greater than 310.

```python
filtered_df = df[df['total_bedrooms'] > 310]
print("Rows where 'total_bedrooms' is greater than 310:")
print(filtered_df)
```

## Drop first and last row

### Subtask:
Drop the first and last rows from the DataFrame.

```python
df = df.iloc[1:-1]
print("DataFrame after dropping first and last rows:")
print("First 5 rows:\n", df.head())
print("\nLast 5 rows:\n", df.tail())
```

```python
mean_households = df['households'].mean()
print(f"Mean of 'households' column: {mean_households}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.scatter(df.index, df['households'], color='blue', label='Households per ID', alpha=0.6)
plt.axhline(y=mean_households, color='red', linestyle='--', label=f'Mean Households: {mean_households:.2f}')
plt.xlabel('ID (DataFrame Index)')
plt.ylabel('Number of Households')
plt.title('Number of Households vs. ID with Mean')
plt.legend()
plt.grid(True)
plt.show()
```

### Subtask:
Check if any of the columns contain NaN values, if yes, replace the NaN values with the arithmetic mean of the whole dataset.

```python
print("Checking for NaN values before replacement:")
print(df.isnull().sum())

# Calculate the arithmetic mean of the entire numerical dataset
# This means calculating the mean of all numerical values in the DataFrame
# We will only consider numerical columns for this operation
numerical_cols = df.select_dtypes(include=['number']).columns

if df[numerical_cols].isnull().any().any():
    dataset_mean = df[numerical_cols].stack().mean()
    df[numerical_cols] = df[numerical_cols].fillna(dataset_mean)
    print(f"\nNaN values replaced with the arithmetic mean of the dataset ({dataset_mean:.2f}).")
else:
    print("\nNo NaN values found in the numerical columns.")

print("\nChecking for NaN values after replacement:")
print(df.isnull().sum())
```

 - display *lat* and *long* into plot, *lat* would be on *x* axis and *long* on *y* axis

```python
plt.figure(figsize=(10, 8))
plt.scatter(df['latitude'], df['longitude'], alpha=0.5, c='purple', s=10)
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Geographical Distribution of Housing Data (Latitude vs. Longitude)')
plt.grid(True)
plt.show()
```

## Min-max normalization

### Subtask:
Select two arbitrary numerical columns from the DataFrame and apply Min-Max Normalization to scale their values to a range between 0 and 1.


```python
col1 = 'total_rooms'
col2 = 'median_income'

# Apply Min-Max Normalization to 'total_rooms'
df[f'{col1}_normalized'] = (df[col1] - df[col1].min()) / (df[col1].max() - df[col1].min())

# Apply Min-Max Normalization to 'median_income'
df[f'{col2}_normalized'] = (df[col2] - df[col2].min()) / (df[col2].max() - df[col2].min())

print(f"DataFrame head with normalized '{col1}' and '{col2}' columns:")
print(df.head())
```
